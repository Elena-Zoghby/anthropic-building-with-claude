````python
import anthropic as anth

client = anth.Anthropic()
model = "claude-sonnet-4-5"
messages = []

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages,system=None,stop=[]):
    params={
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "stop_sequences": stop
        }
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
````

````python
messages=[]
prompt="""
Generate three different sample AWS CLI commands. Each should be
very short.
"""
add_user_message(messages, prompt,stop=["```"])
add_assistant_message(messages,"Here are three sample AWS CLI commands:\n```bash")
answer = chat(messages)
answer.strip()
````
