```python
import anthropic as anth

client = anth.Anthropic()
model = "claude-sonnet-4-5"
messages = []

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)

def chat(messages,system=None):
    params={
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        }
    if system:
        params["system"] = system
    message = client.messages.create(**params)
    return message.content[0].text
```

```python

system_prompt = """
You're a python developer. Provide concise, correct, and short solutions
to the user requests. If the user asks for a code snippet, provide only the code snippet without any explanations or comments.
"""
add_user_message(messages,"Write a python script to remove duplicate from strings")
answer=chat(messages,system_prompt)
answer

```
